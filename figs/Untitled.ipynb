```python
import pandas as pd
import numpy as np

%matplotlib inline

from DataAnalysis.Histograms import Histogram
```

```python
ch0Name = "channel_0.txt"
ch2Name = "channel_2.txt"
ch4Name = "channel_4.txt"
```

```python
# Read in Meulders data and normalize
ch0Data = pd.read_table(ch0Name, delim_whitespace=True, names=['eCh', 'counts'])
ch2Data = pd.read_table(ch2Name, delim_whitespace=True, names=['eCh', 'counts'])
ch4Data = pd.read_table(ch4Name, delim_whitespace=True, names=['eCh', 'counts'])

# Build histogram objects
ch0Histo=Histogram()
ch0Histo.build_histo(ch0Data['eCh'].tolist(), ch0Data['counts'].tolist(), edgeLoc='low', name='Channel 0')
ch2Histo=Histogram()
ch2Histo.build_histo(ch4Data['eCh'].tolist(), ch2Data['counts'].tolist(), edgeLoc='low', name='Channel 2')
ch4Histo=Histogram()
ch4Histo.build_histo(ch4Data['eCh'].tolist(), ch4Data['counts'].tolist(), edgeLoc='low', name='Channel 4')


ch0Histo.plot(ch2Histo, ch4Histo, logY=True, xMax = 120000, includeMarkers=False, xLabel='Channel Number',
              yLabel='Counts', color=['k', 'b', 'g'], dashes=[[10, 0.001], [10, 0.001], [10, 0.001]],
             savePath='uncal-phs-allCh')
ch0Histo.plot(ch2Histo, ch4Histo, logY=True, xMin=0, xMax = 1500, includeMarkers=False, xLabel='Channel Number',
              yLabel='Counts', color=['k', 'b', 'g'], dashes=[[10, 0.001], [10, 0.001], [10, 0.001]],
             savePath='uncal-phs-allCh-zoomed')
```
